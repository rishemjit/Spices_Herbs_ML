### finding the tags corresponding to every cuisine :  tarla dalal

```python
# Tarla

# CUISINE REGION : CUISINE TAGS
# cuisine category is modified to be the same in cuisine-state excel sheet

# There are multiple tags corresponding to a cuisine such as Bengali : 'Bengali Chutney / Raitas', 'Bengali Dals / Curry'
# cuisinetag_map_tarla -> dict, keys-> cuisine , values-> list of tags associated with partcular cusine

import re 
from string import digits 
import json

read_file = '../Crawlers/TarlaDalal_English/cuisine_tag_tarla.txt'
write_file = result_folder+ 'cuisinetag_map_tarla.json'
with open(read_file) as r_f:
    content = r_f.readlines()
    
cuisinetag_map_tarla = {}
tag_list = []
for line in content:
    
    if line[0].isdigit():
        # update dict with collected category: tag list mapping before move on to the next category
        if tag_list != []:
            cuisinetag_map_tarla.update({category: tag_list})
        # the next category and new tag list 
        category = re.sub('\(.*\)','', line).lstrip(digits).lstrip().rstrip()
        
        if category != 'South Indian':
            # to match the cuisine type in cuisineregion_state excel sheet
            category = category.split()[0]
        tag_list = [category] # this hould be changed to include category as well by rishem
    else:
        tag = re.sub('\(.*\)','',line).lstrip().rstrip()
        if tag == 'Maharashtrian Bhaat': # by rishem
            tag = tag+ ' (Rice)'          # by rishem
        if tag == 'Maharashtrian Upvas':   # by rishem
            tag= tag+ ' (Fasting)'         # by rishem
        tag_list.append(tag)
cuisinetag_map_tarla.update({category: tag_list})

print(cuisinetag_map_tarla,'\n\n # of cuisines: ',len(cuisinetag_map_tarla.keys()))

regiontags_tarla = list(chain.from_iterable(cuisinetag_map_tarla.values()))

dict2json(cuisinetag_map_tarla, write_file)

print(cuisinetag_map_tarla.keys())
```

### tarla dalal: recipes and corresponding tags

```python
# Tarla CUISINE recipe names and corresponding cuisine tags 
# cuisine_tag_dict_tarla -> keys -> recipe names, values-> corresponding cuisine tags (not cuisine)


cuisine_tag_dict_tarla = {}
for i in range(len(recipe_tarla)):
    tag = recipe_tarla[i]['cuisine']
    if tag != []:
        cuisine_tag_dict_tarla.update({recipe_tarla[i]['title']['text'] : tag})
    else:
        print('recipe without tags ',recipe_tarla[i]['title']['text'] )
        
dict2json(cuisine_tag_dict_tarla, result_folder + 'cuisine_tag_dict_tarla.json')

print('\n no of recipes with cuisine tags , tarla dalal', len(cuisine_tag_dict_tarla))

dict(islice(cuisine_tag_dict_tarla.items(), 4))  
```

### tarla dalal: recipes and corresponding cuisine name
** Please note that a recipe can belong to more than one cuisine e.g. Vegetable Corn Bake ( Diabetic Recipe )': ['Jain', 'Gujarati'],

```python
#cuisinerecipe2region_tarla -> dict, keys-> name of recipe, values-> list of name of cuisines
# one recipe can be associated with multiple cuisines

tag_notpresent = []
cuisine_regiontag_dict_tarla = {}
for cuisine, tags in cuisine_tag_dict_tarla.items():
    regiontag_list = []
    for t in tags:
        t = re.sub(' +', ' ', t).rstrip() ## by rishem 
        if t in regiontags_tarla:
            regiontag_list.append(t)
        else:
            
            tag_notpresent.append(t)  # by rishem
            
    if regiontag_list:
        cuisine_regiontag_dict_tarla.update({cuisine: regiontag_list})
dict2json(cuisine_regiontag_dict_tarla, result_folder+'cuisine_regiontag_dict_tarla.json')


print('tags not found ', len(set(tag_notpresent)), set(tag_notpresent)   )


cuisinerecipe2region_tarla = {}
for cuisine, regiontags in cuisine_regiontag_dict_tarla.items():
    region_list = []
    for regiontag in regiontags:
        for region in cuisinetag_map_tarla.keys():
            if regiontag in cuisinetag_map_tarla[region]: 
                region_list.append(region)
    # remember to drop the duplicate regions (there may be multiple region tags point to same region)
    cuisinerecipe2region_tarla.update({cuisine : list(set(region_list))}) 
    
region2cuisinerecipe_tarla = reverse_dict(cuisinerecipe2region_tarla) #shuchen
dict2json(region2cuisinerecipe_tarla, result_folder+'region2cuisinerecipe_tarla.json') #shuchen
region2recipecount_tarla = dict2countdict(region2cuisinerecipe_tarla, region2cuisinerecipe_tarla.keys()) #shuchen
dict2json(region2recipecount_tarla, result_folder+'region2recipecount_tarla.json') #shuchen

print('\n no of recipes in tarla Dalal after cleaning region tags', len(cuisinerecipe2region_tarla))

dict(islice(cuisinerecipe2region_tarla.items(), 4))  
```

```python
#cuisinerecipe2region_sanj -> dict, keys-> name of recipe, values-> list of name of cuisines
# cuisine2recipecount_sanj -> dict, keys-> cuisine name, values-> count of recipes
# cuisine_overlapcount_sanj-> dataframe


cuisine2recipecount_sanj = dict.fromkeys(cuisines_sanj, 0)
cuisinerecipe2region_sanj = {}
tag_notpresent = []

for i in range(len(recipe_sanj)):
    cuisine = recipe_sanj[i]['title']['text']
    country = recipe_sanj[i]['country']
    region_list = []
    if country == 'IN':           # if country is India
        tag = recipe_sanj[i]['tags']['cuisine']
        #print( country, tag, cuisine)
        if tag != []:
            for t in tag:
                temp = t.rstrip().lstrip()
                if temp in tag_sanj:
                    region_list.append(tagcuisine_map_sanj[temp])
                    cuisine2recipecount_sanj[tagcuisine_map_sanj[temp]] +=1
                else:
                    tag_notpresent.append(temp)
                                                
            if region_list != []:
                cuisinerecipe2region_sanj.update({cuisine : list(set(region_list))})
            
dict2json(cuisine2recipecount_sanj, result_folder+'region2recipecount_sanj.json') #shuchen
region2cuisinerecipe_sanj = reverse_dict(cuisinerecipe2region_sanj) #shuchen
dict2json(region2cuisinerecipe_sanj, result_folder+'region2cuisinerecipe_sanj.json') #shuchen

print('no of recipes in Sanjeev Kapoor belonging to Indian cusines', len(cuisinerecipe2region_sanj))

print('\n tags not found: ', set(tag_notpresent))

print('\n', cuisine2recipecount_sanj)
dict(islice(cuisinerecipe2region_sanj.items(), 4))
cuisine2recipecount_sanj
```

```python
#cuisine_overlapcount_sanj -> dataframe

import matplotlib.pyplot as plt
cuisine_overlapcount_sanj = pandas.DataFrame(0, columns=cuisines_sanj, index=cuisines_sanj )
no_region= []

for recipe, region_list in cuisinerecipe2region_sanj.items():
    temp= len(region_list)
    no_region.append(temp)
    if temp == 2:
        cuisine_overlapcount_sanj[region_list[0]][region_list[1]]+=1
        cuisine_overlapcount_sanj[region_list[1]][region_list[0]]+=1
    elif temp ==3:
        cuisine_overlapcount_sanj[region_list[0]][region_list[1]]+=1
        cuisine_overlapcount_sanj[region_list[1]][region_list[2]]+=1
        cuisine_overlapcount_sanj[region_list[0]][region_list[2]]+=1
        cuisine_overlapcount_sanj[region_list[1]][region_list[0]]+=1
        cuisine_overlapcount_sanj[region_list[2]][region_list[1]]+=1
        cuisine_overlapcount_sanj[region_list[2]][region_list[0]]+=1
    else:
        cuisine_overlapcount_sanj[region_list[0]][region_list[0]]+=1
#print(cuisine_overlapcount_sanj)
temp = plt.hist(no_region)
plt.title('Sanjeev Kapoor')
plt.show()
print('fraction of recipes associated with single cuisine', temp[0][0]/sum(temp[0]))
```

## find spices present in tarla dalal and sanjeev kapoor after cleaning. 

```python
# SPICE RECIPE - SPICE (HANDBOOK) 
# Mapping between spice (appeared in recipes) and spice/herb ingre(appeared in handbooks)
 
# spice recipe data excel modified after compared to tarla and sanjeev 
# bay leaves or bay leaf 
# pigeon pea (if contain pegeon pea then include)
# roasted coconut (if contain coconut then include)
# couldn't find cassia, chinese 5 spice powder, sage, sandwich masala, marjoram, coloured aniseed sweets,black mustard seed oil in tarla or sanjeev

import unidecode

write_file = result_folder + 'spicetarla_spicehnd_dict.json'
spice_excel = pandas.read_excel('../Crawlers/OtherData/recipespice_data1.xlsx')   # remove 'corainder' since it doesn't occur in the handbook
spice_list = spice_excel.values.tolist()

spice_dict = {}

for line in spice_list: 
    if type(line[1]) is not float and line[1] !='NS': # if not 'NS' and 'NA'; line[1] is entire unicode string
        l = re.sub('(NA_|NS_)[^,]*,','',line[1]) # remove 'NA_','NS_' prefix from the string 
        l = [re.sub('  ',' ',item.lstrip().rstrip().lower()) for item in l.split(',')] # convert string of items to list of items and remove leading spaces
        if(l == []): # if item empty, not insert into dict
            continue
        s = re.sub('  ',' ',line[0]).strip()
        #print(s,l)
        spice_dict.update({s.lower():l})
dict2json(spice_dict, write_file)   

spices_hnd = set(list(chain.from_iterable(spice_dict.values()))) # hndbook spice ingredients in spice recipe
spice_excel= list(spice_dict.keys())  # spices appeared in Tarla Dalal recipe data  

print('Number of Spices in tarla dalal (manual annotation): ', len(spice_excel)) # 137
print('Number of related Spices (handbook): ', len(spices_hnd)) # 108
```

#### Tarlal dalal : finds recipes and corresponding spices




```python
# spices present in recipes
# RECIPE name : SPICEs(TARLA DALA)
# Exclude non-spice ingredients (spice recipe lower case, single space)
#cuisine_dict -> dict,keys-> recipe name, values-> list of spices
# writes tarla_recipename2spice.json with recipe names and spices

write_file = result_folder + 'tarla_recipename2spice.json'
cuisine_dict = cuisine_dict_orig.copy()
for cuis in cuisine_dict_orig.keys():
    ingr_list = []
    for ingr in cuisine_dict_orig[cuis]:
        
        
        #if ' '.join(ingr.lower().split()) in spice_excel:
        index= np.where(list(map(' '.join(ingr.lower().split()).__contains__, spice_excel)))[0] # by rishem, map string with list of substrings
        
                                                                  
        if len(index)>0:
            #print(ingr, index,  list(map(spice_excel.__getitem__, index) ))
            ingr_list.extend(list(map(spice_excel.__getitem__, index) ))# by rishem
            
            
            #ingr_list.append(' '.join(ingr.lower().split()))
        if 'coconut' in ingr.lower():
            ingr_list.append('coconut')
        if 'pigeon pea' in ingr.lower():
            ingr_list.append('pigeon pea')
        if 'pomegranate' in ingr.lower():
            ingr_list.append('pomegranate')
    cuisine_dict[cuis] = list(set(ingr_list))
    if ingr_list == []:   # Remove the  recipe that doesn't contain spices
        #print(cuisine_dict_orig[cuis])
        cuisine_dict.pop(cuis)

j = json.dumps(cuisine_dict,indent = 4)
with open(write_file, 'w') as w_file:
    w_file.write(j)
    
spice_tarla = set(list(chain.from_iterable(cuisine_dict.values()))) 

print('# of  recipes contain spices in Tarla:', len(cuisine_dict.keys()))
print('# of spice  in Tarla: ', len(spice_tarla)) # 84 from tarla
print('# of spice  in total: ', len(spice_excel)) # 109
#print('\nSpice recipes missed from the entire spice recipe set for Tarla: \n', set(spice_excel) - set(spice_tarla))

print(spice_tarla)
```
