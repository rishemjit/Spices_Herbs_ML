```python
# NEED TO SPECIFY cookie each time 

'''
search_indications_mala:
Args:
    query_item (string): indication name
Return:
    content (bs): all search/scraping results
'''

def search_indications_mala(query_item):
    ga = 'GA1.2.318950283.1587144572'
    #gat = '1'
    gid = 'GA1.2.429160416.1587144572'
    incap_ses_113_161858 = 'ygK3B5erKSs+mDM693WRAXjnmV4AAAAAA+HdcCnCwMCCSStxIpSB6A=='
    nlbi_161858 = '65u7FPkdcTEOGx7N1XDSAgAAAAAhARnFxIAWx70Ft66mxow+'
    visid_incap_161858 = 'eASkTK9+T1SOnzec+Xij+vsfel4AAAAAQkIPAAAAAABHyEgQ68g2TNEhDNh1T87O'

    cookie = '_ga='+ga+\
            '; _gid='+gid+\
            '; incap_ses_113_161858='+incap_ses_113_161858+\
            '; nlbi_161858='+nlbi_161858+\
            '; visid_incap_161858='+visid_incap_161858

    headers = {
               'Cookie':cookie, 
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

    #url = 'https://www.malacards.org/'
    qstr = quote(query_item)
    url = 'https://www.malacards.org/search/results?query='+qstr

    for i in range(0,20):
        req=requests.get(url,timeout =5, headers =headers)
        if req.status_code == 200:
            soup = BeautifulSoup(req.text,"html.parser") 
            content = soup.find('div',attrs={'class': 'search-page','data-query':query_item})
            return content
            break    
        else:
            print(i, req.status_code)
    return
```

```python
'''
simlar_indications_mala: scrape top 15 similar mala indications
Args:
    content(bs): all search/scraping results
Return:
    result_list(string list): list of top 15 similar mala indications
'''
import re 
def similar_indications_mala(content, NUM):
    #print(type(content))
    #print(content)
    td_ID = content.findAll('td',attrs={'title':"MalaCards ID"})
    #print(td_ID)
    #print(td_ID)
    result_list = []
    if(len(td_ID) < NUM ):
        num_similars = len(td_ID)
    else:
        num_similars = NUM
    for i in range(num_similars):
        tr = content.find('tr',attrs={'id':re.compile("^" + td_ID[i].contents[0])})
        tr_a = tr.findAll('a')
        if(len(tr_a) == 2):
            result_list.append(tr_a[1].contents[0])
        else:
            result_list.append(tr_a[0].contents[0])
    return result_list
```

```python
def process_name(s):
    return s.lower().replace(' ', '').replace(',', '')
```

```python
disease_cats_malacard = json2dict('..\Crawlers\Malacards\disease_cats_malacard.json')
indic2category_mala = reverse_dict(disease_cats_malacard)
hndindic_processed = [process_name(indic) for indic in hndindic_all]
mala_indic = set(list(chain.from_iterable(disease_cats_malacard.values())))  
mala_indic_processed =  [process_name(indic) for indic in mala_indic]
```

```python
from collections import Counter
data = []
hndindic2match_top1 = {}
hndindic_missed = []
count_exactmatch = 0 # count exact match
count_miss = 0
c = 0
for indic, match in hndindic2match.items():
    category1 = ''
    category2 = ''
    top_match = ''
    if match:  
        # if top 1 is the exact match
        if match[0].lower() == indic.lower():
            count_exactmatch += 1
        top_match = ''
        top_match = match[0]
        # if top match not appear in mala category, use next top match
        if top_match not in indic2category_mala:
            for i in range(1, num_match):
                if match[i] in indic2category_mala:
                    top_match = match[i]
                    break
        if top_match in indic2category_mala:
            if len(indic2category_mala[top_match]) > 1:
                categories = []
                for m in match:
                    if m in indic2category_mala:
                        categories.extend(indic2category_mala[m])
                categories_count = Counter(categories)
                max_count = max(categories_count.values())
                categories_top = [k for k in categories_count if categories_count[k] == max_count] 
                category1 = categories_top[0].replace('_list', '')
                if len(categories_top) > 1:
                    c += 1
                    category2 = categories_top[1].replace('_list', '')
                    #print(indic, categories_count)
        else:
            print('NOT FOUND', indic)      
    else: # if there's no match through query
        hndindic2match_top1[indic] = ''
        indic_processed = process_name(indic)
        # check if there's match in mala indic list
        if indic_processed in mala_indic_processed:
            print(indic)
        else:
            # potential matches through substring
            for malaindic in mala_indic_processed:
                if indic_processed in malaindic:
                    print(indic, '(hnd) = ? ', malaindic, '(mala)')
            hndindic_missed.append(indic)
        count_miss += 1
    data.append({'hnd_indic': indic,
                 'mala_indic': top_match,
                 'category1': category1,
                 'category2': category2})
dict2json(hndindic2match, 'Handbook_Data/hndindic2match_top' + str(1) + '.json')
print('num exact match:', count_exactmatch)
print('num hndindic missing matches:', count_miss)
```

```python
indic2match2category_df = pd.DataFrame(data, columns=["hnd_indic", "mala_indic", "category1", "category2"])
indic2match2category_df.to_excel('Handbook_Data/indic_all_categorized_malacards.xlsx')
```
