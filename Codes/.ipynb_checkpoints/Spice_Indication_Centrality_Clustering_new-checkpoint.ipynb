## To analyze the clusters obtained

```python
# add cluster column to Spice_Indication_df
def df_spice2indic_clustered(cluster_pd, Spice_Indication_df):
    print(len(cluster_pd.index))
    for i in range(len(cluster_pd.index)):
        spice = cluster_pd.loc[i, 'Vertex']
        cluster = cluster_pd.loc[i, 'Group']
        if 'G' not in cluster:
            print(spice)
        Spice_Indication_df.loc[spice, 'Cluster'] = cluster
    return Spice_Indication_df
spice2indic_clustered_wakita = df_spice2indic_clustered(cluster_pd, Spice_Indication_df)
spice2indic_clustered_wakita.head()
```

### delete clusters with less than 20 spices(can be experimented)

```python
def df_cluster2count(Spice_Indication_df):
    temp = Spice_Indication_df['Cluster'].value_counts() # find no of spices in each cluster
    print(temp)
    # delete cluster with less than 20 enteries
    # for each cluster, each indication receive score calculated by # associated spices in cluster / # spices in cluster 
    group_ind_count =Spice_Indication_df.groupby('Cluster').apply(lambda x: x.sum()/len(x))
    keepClust = temp.loc[temp>5].index
    delClust = set(temp.index ) - set(keepClust)
    group_ind_count.drop(delClust, inplace=True)
    return group_ind_count
cluster2count_wakita = df_cluster2count(spice2indic_clustered_wakita)
cluster2count_wakita
```

### Clustering 2: Leidenalg Clustering ###

```python
# generate adjacency matrix of nx projection graphs (pd format or np arr)

from pandas import ExcelWriter
import igraph as ig
import leidenalg
import os
```

```python
if IDX == 0:
    adjmtx_np = nx.convert_matrix.to_numpy_matrix(top_projection)
    adjmtx_pd = nx.to_pandas_adjacency(top_projection)

    adjmtx_np_backbone = nx.convert_matrix.to_numpy_matrix(top_backbone_nx)
    adjmtx_pd_backbone = nx.to_pandas_adjacency(top_backbone_nx)
else:
    adjmtx_np = nx.convert_matrix.to_numpy_matrix(bot_projection)
    adjmtx_pd = nx.to_pandas_adjacency(bot_projection)

    adjmtx_np_backbone = nx.convert_matrix.to_numpy_matrix(bot_backbone_nx)
    adjmtx_pd_backbone = nx.to_pandas_adjacency(bot_backbone_nx)

node_names = list(adjmtx_pd.columns.values)

graph_pd = pd.DataFrame(adjmtx_np, index=node_names, columns=node_names)

# Get the values as np.array, it's more convenenient.
A = graph_pd.values

# Create graph, A.astype(bool).tolist() or (A / A).tolist() can also be used.
graph_ig = ig.Graph.Adjacency((A > 0).tolist())

# Add edge weights and node labels.
graph_ig.es['weight'] = A[A.nonzero()]
graph_ig.vs['label'] = node_names  # or a.index/a.columns
```

```python
PART_IDX = 0
node2cluster_file = 'NODEXL_PLOT_NEW/' + PROJ_MODE[0] + '/'+ PROJ_MODE[IDX] + '_cluster2node'+'_leiden_' + str(PART_IDX) + '.xlsx'
if not os.path.isfile(node2cluster_file):
    # install igraph and leidenalg
    # search igraph graph from numpy or pandas adjacency matrix
    RESOLUTION = 0.1

    parts_modularity = leidenalg.find_partition(graph_ig, leidenalg.ModularityVertexPartition, n_iterations=-2) #0

    parts_significance = leidenalg.find_partition(graph_ig, leidenalg.SignificanceVertexPartition) #1

    parts_RB = leidenalg.find_partition(graph_ig, leidenalg.RBConfigurationVertexPartition, resolution_parameter = RESOLUTION) #2 

    parts_RBER = leidenalg.find_partition(graph_ig, leidenalg.RBERVertexPartition, resolution_parameter = RESOLUTION) #3 

    parts_surprise = leidenalg.find_partition(graph_ig, leidenalg.SurpriseVertexPartition) #4 

    parts_CPM = leidenalg.find_partition(graph_ig, leidenalg.CPMVertexPartition, resolution_parameter = RESOLUTION) #5

    parts_list = [parts_modularity, parts_significance, parts_RB, parts_RBER, parts_surprise, parts_CPM]

    # generate {node name: group id} dict for entire projection graph

    node2group = {}
    partition = parts_list[PART_IDX]
    for group_id in range(len(partition)):
        for i in partition[group_id]:
            node2group[node_names[i]] = 'G' + str(group_id + 1)
    cluster_pd_leiden = pd.DataFrame(node2group.items(), columns=['Vertex', 'Group'])
    print(cluster_pd_leiden)
    cluster_pd_leiden.to_excel(node2cluster_file)
    spice2indic_clustered_leiden = df_spice2indic_clustered(cluster_pd_leiden, Spice_Indication_df)
    spice2indic_clustered_leiden.head()
else:
    #print('READ---')
    cluster_pd_leiden = pd.read_excel(node2cluster_file)
    #print(cluster_pd_leiden
    spice2indic_clustered_leiden = df_spice2indic_clustered(cluster_pd_leiden, Spice_Indication_df)
spice2indic_clustered_leiden.head()
```

### Centrality Measures ###

```python
# Centrality

from collections import OrderedDict
from operator import itemgetter    

if IDX == 0:
    g_distance_dict = {(e1, e2): 1 / weight for e1, e2, weight in top_projection.edges(data='weight')}
    nx.set_edge_attributes(top_projection, g_distance_dict, 'distance')
        
    centrality_eigen = nx.eigenvector_centrality(top_projection, weight='weight')

    #centrality_katz = nx.katz_centrality(GRAPHS[GRAPH_IDX])

    #centrality_harmonic = nx.harmonic_centrality(top_projection)

    centrality_closeness = nx.closeness_centrality(top_projection, distance='distance')

    #centrality_degree = nx.degree_centrality(top_projection)
    centrality_degree = top_projection.degree(weight='weight')
else:
    g_distance_dict = {(e1, e2): 1 / weight for e1, e2, weight in bot_projection.edges(data='weight')}
    nx.set_edge_attributes(bot_projection, g_distance_dict, 'distance')
    centrality_eigen = nx.eigenvector_centrality(bot_projection, weight='weight')

    #centrality_katz = nx.katz_centrality(GRAPHS[GRAPH_IDX])

    centrality_harmonic = nx.harmonic_centrality(bot_projection)

    centrality_closeness = nx.closeness_centrality(bot_projection)

    #centrality_degree = nx.degree_centrality(bot_projection)
    centrality_degree = bot_projection.degree(weight='weight')
```

```python
centrality_eigen_ordered = list(OrderedDict(sorted(centrality_eigen.items(), key = itemgetter(1), reverse = True)))
print('eigen_centrality: ', centrality_eigen_ordered)

centrality_closeness_ordered = list(OrderedDict(sorted(centrality_closeness.items(), key = itemgetter(1), reverse = True)))
print('\ncloseness centrality: ',centrality_closeness_ordered)

centrality_degree_ordered = list(OrderedDict(sorted(centrality_degree, key = itemgetter(1), reverse = True)))
print('\ndegree centrality: ', centrality_degree_ordered)
```

```python
# combine results and find top 20
TOP = 20
spice2rank = {}
for spice in centrality_closeness_ordered:
    rank_eigen = centrality_eigen_ordered.index(spice)
    rank_closeness = centrality_closeness_ordered.index(spice)
    rank_degree = centrality_degree_ordered.index(spice)
    rank_combined = rank_eigen + rank_closeness + rank_degree
    spice2rank.update({spice : rank_combined})
spice_ranked = list(OrderedDict(sorted(spice2rank.items(), key = itemgetter(1), reverse = False)))
', '.join(spice_ranked[:TOP])
```

```python
df = pd.DataFrame.from_dict({'Eigen Centrality':centrality_eigen_ordered,'Closeness Centrality': centrality_closeness_ordered,
                             'Degree Centrality': centrality_degree_ordered, 'Combined rank': spice_ranked })
df.to_excel('NODEXL_PLOT_NEW/' + PROJ_MODE[0] + '/' + PROJ_MODE[IDX] +  '_centrality_summary.xlsx', header=True, index=False)
```
