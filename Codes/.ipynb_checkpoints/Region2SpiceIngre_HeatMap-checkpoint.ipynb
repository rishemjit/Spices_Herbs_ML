```python
def region2spiceingre_heatmap(region2recipecount, region2cuisinerecipe, cuisinerecipe2spiceingre):
    region2spiceingre = {}
    for region, cuisine_list in region2cuisinerecipe.items():
        spiceingre_list = []  # list of spiceingre per region
        for cuisine in cuisine_list: 
            if cuisine in cuisinerecipe2spiceingre.keys():  # some cuisinerecipe occur in region2cuisinerecipe dict
                                                            #    but not cuisinerecipe2spiceingre dict
                spiceingre_list.extend(cuisinerecipe2spiceingre[cuisine]) 
        count_dict = Counter(spiceingre_list) # count dict of spiceingre per region
        for spice in count_dict.keys():
            count_dict[spice] = count_dict[spice]/region2recipecount[region]
        region2spiceingre.update({region : count_dict}) 
    return region2spiceingre

# Sanj
region2spiceingre_sanj_hm = region2spiceingre_heatmap(region2recipecount_sanj, region2cuisinerecipe_sanj, 
                                                      cuisinerecipe2spiceingre_sanj)

# Tarla
region2spiceingre_tarla_hm = region2spiceingre_heatmap(region2recipecount_tarla, region2cuisinerecipe_tarla, 
                                                       cuisinerecipe2spiceingre_tarla)


# we will add random recipe data set as well in the combined data

if include_random == True:
    #randomrecipe2spiceingre = json2dict('Derived_Data/CopyMutate/random_recipe_uniform.json')
    randomrecipe2spiceingre = json2dict('Derived_Data/CopyMutate/.json')

    cuisinerecipe2spiceingre_combined.update(randomrecipe2spiceingre) # append to Indian cuisine dictionary
    region2recipecount_combined['Random'] = len(randomrecipe2spiceingre)
    region2cuisinerecipe_combined['Random'] = list(randomrecipe2spiceingre.keys())

# Combined 
region2spiceingre_combined_hm = region2spiceingre_heatmap(region2recipecount_combined, region2cuisinerecipe_combined, 
                                                          cuisinerecipe2spiceingre_combined)
```

```python
# Plot heatmap for region-spiceingre mapping for both Tarla and Sanjeev



def plot_heatmap(val_arr,idx_list, col_list, cmap, title, savefig_name):
    plt.figure(figsize=(16,10))
    ax = plt.gca()
    im = ax.imshow(val_arr, cmap = cmap)

    ax.set_xticks(np.arange(len(idx_list)))
    ax.set_yticks(np.arange(len(col_list)))
    print(len(idx_list))
    print(len(col_list))
    # label with the respective list entries
    ax.set_xticklabels(idx_list)
    ax.set_yticklabels(col_list)

    # Rotate the tick labels and set alignment.
    plt.setp(ax.get_xticklabels(), rotation=90, ha="right",
             rotation_mode="anchor")
    ax.set_title(title)

    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="2%", pad=0.08)

    plt.colorbar(im, cax=cax)
    plt.show()
    #plt.savefig(result_folder+ savefig_name)

plot_heatmap(region2spiceingre_df_sanj.values.T, list(region2spiceingre_df_sanj.index) , list(region2spiceingre_df_sanj.columns), 'jet', 'Cuisine Style - Spice Ingredient Heat Map (based on Sanjeev dataset)', 'region2spiceingre_heatmap_sanj.png')
plot_heatmap(region2spiceingre_df_tarla.values.T, region2spiceingre_df_tarla.index , region2spiceingre_df_tarla.columns, 'jet', 'Cuisine Style - Spice Ingredient Heat Map (based on Tarladalal dataset)', 'region2spiceingre_heatmap_tarla.png')
plot_heatmap(region2spiceingre_df_combined.values.T, list(region2spiceingre_df_combined.index) , list(region2spiceingre_df_combined.columns), 'jet', 'Cuisine Style - Spice Ingredient Heat Map (based on Tarla + Sanjeev)', 'region2spiceingre_heatmap_combined.png')
```

```python


#plotvar = region2spiceingre_df_combined.T


def myplot(score,coeff,label_sample, labels=None, dist = 'euclidean'):
    from scipy.spatial.distance import cdist
    pc_ob=[0,1]
    print('score shape', score.shape)
    xs = score[:,pc_ob[0]]
    ys = score[:,pc_ob[1]]
    n = coeff.shape[0]
    #scalex = 1.0/(xs.max() - xs.min())
    #scaley = 1.0/(ys.max() - ys.min())
    #plt.scatter(xs * scalex,ys * scaley)
    plt.scatter(xs ,ys )
    dist = cdist(coeff[:,pc_ob], [(0,0)], dist)
    distcomp= np.sort(dist, axis=0)[::-1][20]# plot top 30 loadings
    textall=[]
    #print(n)
    for i in range(n):
        if dist[i]> distcomp:
            plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = [0.1, 0.2, 0.3],alpha = 0.5)
            if labels is None:
                plt.text(coeff[i,0]* 1.0, coeff[i,1] * 1.0, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
            else:
                textall.append(plt.text(coeff[i,pc_ob[0]]* 1.0, coeff[i,pc_ob[1]] * 1.0, labels[i], color = 'g', ha = 'center',
                                        va = 'center', 
                         fontsize=12))
    #adjust_text(textall)
    
    #textall=[]

    for i in range( len(label_sample)):
        textall.append(plt.text(xs[i]*1, ys[i]*1, label_sample[i], fontsize=14))
    adjust_text(textall)
    #plt.xlim(-1,1)
    #plt.ylim(-1,1)
    print (pca.explained_variance_ratio_)
    plt.xlabel("PC{} ({:.2f}%)".format(pc_ob[0]+1, pca.explained_variance_ratio_[0]*100), fontsize= 14)
    plt.ylabel("PC{} ({:.2f}%)".format(pc_ob[1]+1, pca.explained_variance_ratio_[1]*100), fontsize=14)
    plt.grid()
    #plt.savefig(result_folder+ 'PCA_cuisine_spices_combined.png')
    
    
    
```

```python

# find top spices for each cuisine


def plot_top_heatmap(region2ind,k, cmap= sns.cm.rocket_r):
    top_ind_list = []
    for i, ind in enumerate(region2ind.index):
        t = region2ind.loc[ind].sort_values(ascending = False)[:k]
        top_ind_list.extend(list(t.index))

    cluster_ind_df = pd.DataFrame(index=region2ind.index,  columns = list(set(top_ind_list)), dtype='float64')

    for i, ind in enumerate(region2ind.index):
        cluster_ind_df.loc[ind] = region2ind.loc[ind, cluster_ind_df.columns]

    #fig = plt.figure(figsize= (10,5))
    sns.heatmap(cluster_ind_df, cmap= cmap)



def plot_prevalence(region2spiceingre_df_combined, no_subplot=[6,4]):
    fig = plt.figure(figsize= (20,20))
    fig.tight_layout()
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)


    for i, ind in enumerate(region2spiceingre_df_combined.columns):
        t = region2spiceingre_df_combined[ind].sort_values(ascending = False)[:10] # find top spices
        #t = t[t>0.5]
        ax = fig.add_subplot(no_subplot[0], no_subplot[1], i+1)
        t.plot.bar(ax=ax)
        #temp = Spice_Indication_df.loc[Spice_Indication_df['Cluster']==ind].index
        #list (Spice_Indication_df.loc[temp].sum(axis=1).sort_values(ascending=False)[:10].index)

        plt.title(ind)
plot_prevalence(region2spiceingre_df_combined)

plt.figure(figsize =[10,8])
plot_top_heatmap(region2spiceingre_df_combined.T,15)
plt.savefig(result_folder+ 'prevalence_topspices.png', bbox_inches = 'tight')
```

### find relative prevalence

```python
## relative prevalence 

def cal_relative_prev(region2spiceingre_df_combined):
    no_cuisine = len(region2spiceingre_df_combined.columns)

    avg = region2spiceingre_df_combined.sum(axis=1)

    temp = region2spiceingre_df_combined*no_cuisine
    relative_prev = temp.sub(avg, axis=0)/(no_cuisine-1)
    return relative_prev

relative_prev= cal_relative_prev(plotvar)
plot_prevalence(relative_prev)
```

### authenticity of spices pairs

```python
def cal_pairing(region2cuisinerecipe_combined,region2recipecount_combined, k=2):

    ingrepair_cuisine = pd.DataFrame(columns= region2cuisinerecipe_combined.keys()) 

    for cuisine, recipe_list in region2cuisinerecipe_combined.items():
        for recipe in recipe_list:
            ing_list = cuisinerecipe2spiceingre_combined[recipe]
            for comb in list(itertools.permutations(ing_list, k)):
                #print(comb, comb[::-1])
                comb_reverse = str(comb[::-1]) #11june, 2020 to remove the duplicat epair
                comb = str(comb)

                #print(ingrepair_cuisine.index)
                #print(comb_reverse)
                if comb_reverse in ingrepair_cuisine.index:
                    ingrepair_cuisine.loc[comb_reverse, cuisine]+=1
                    #print('exists',comb_reverse)
                elif comb in ingrepair_cuisine.index:
                    ingrepair_cuisine.loc[comb, cuisine]+=1
                   # print('exists', comb)
                #if comb in ingrepair_cuisine.index or comb_reverse in ingrepair_cuisine.index: #11june, 2020 
                    #ingrepair_cuisine.loc[comb, cuisine]+=1
                else:
                    ingrepair_cuisine.loc[comb] =0
                    ingrepair_cuisine.loc[comb, cuisine]=1
                    #print('new added', comb)
        ingrepair_cuisine[cuisine]= ingrepair_cuisine[cuisine]/region2recipecount_combined[cuisine]
    return ingrepair_cuisine

        
            
ingrepair_cuisine = cal_pairing(region2cuisinerecipe_combined,region2recipecount_combined, 2)       
```
