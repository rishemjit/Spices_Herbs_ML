```python

def region2spiceingre_heatmap(region2recipecount, region2cuisinerecipe, cuisinerecipe2spiceingre):
    region2spiceingre = {}
    for region, cuisine_list in region2cuisinerecipe.items():
        spiceingre_list = []  # list of spiceingre per region
        for cuisine in cuisine_list: 
            if cuisine in cuisinerecipe2spiceingre.keys():  # some cuisinerecipe occur in region2cuisinerecipe dict
                                                            #    but not cuisinerecipe2spiceingre dict
                spiceingre_list.extend(cuisinerecipe2spiceingre[cuisine]) 
        count_dict = Counter(spiceingre_list) # count dict of spiceingre per region
        for spice in count_dict.keys():
            count_dict[spice] = count_dict[spice]/region2recipecount[region]
        region2spiceingre.update({region : count_dict}) 
    return region2spiceingre
```

```python
# Combined recipes
# READ COMBINED DICT

# cuisine_spicehnd_dict_combined.json contains spice names as mapped in the handbook
#cuisinerecipe2spicerecipe_combined.json contains spice names as in the tarla dalal

cuisinerecipe2spiceingre_combined = json2dict('Derived_Data/Overlap/cuisine_spicehnd_dict_combined.json') 

region2cuisinerecipe_combined = json2dict('Derived_Data/Overlap/region2cuisinerecipe_combined.json')
cuisinerecipe2region_combined = reverse_dict(region2cuisinerecipe_combined)
region2recipecount_combined = dict2countdict(region2cuisinerecipe_combined, region2cuisinerecipe_combined.keys())
```

```python

# Combined 
region2spiceingre_combined_hm = region2spiceingre_heatmap(region2recipecount_combined, region2cuisinerecipe_combined, 
                                                          cuisinerecipe2spiceingre_combined)

# Combined 
region2spiceingre_df_combined = pd.DataFrame(region2spiceingre_combined_hm).fillna(0)
region2spiceingre_df_combined.sort_index(ascending=True)

region2spiceingre_df_combined = region2spiceingre_df_combined.reindex(sorted(region2spiceingre_df_combined.columns), axis=1)
```

```python
## read spice2indication file. this consists of spice names as in handbook
spice2ind = json2dict("PHYTOCHEMICALS/spiceherb_indication.json") 

spice_cuisine = region2spiceingre_df_combined.index # names of spices present in Indian cuisines

print('no of spices/herbs in handbook and phytochemicals {}, no of spices in Indian cuisine {}'.format(
    len(spice2ind.keys()), len(spice_cuisine)))

spice_cuisine_found = set(spice2ind.keys()) & set(spice_cuisine)

print('overlap between cuisine spice and spice handbook', len(spice_cuisine_found))
print('Indian spices not found in handbook', set(spice_cuisine) -spice_cuisine_found)

```

```python
cuisine_spice_dict = {} # dict with cuisine names as keys and list of spices present as values
cuisine_spicecost_dict= {}



for cuisine in cuisine_names:
    temp = region2spiceingre_df_combined.loc[region2spiceingre_df_combined[cuisine]>0, cuisine]
    spice_list =temp.index.tolist()
    spice_dict = {}
    cost_dict = {}
    for spice in spice_list:
        if spice in spice_cuisine_found:
            spice_dict[spice] = list(spice2ind[spice].keys())
            cost_dict[spice] = region2spiceingre_df_combined.loc[spice, cuisine ]
    cuisine_spice_dict[cuisine] = spice_dict
    cuisine_spicecost_dict[cuisine] = cost_dict
```

### min set cover for each cuisine

```python


results_pd = pd.DataFrame(index = cuisine_names, columns= ['#Spices', '#Indications', 'Min number of spices to cover indications', 'Min set'])


indic2spice_adjmtx = []
for group in cuisine_names:
    set_name = ['Spices', 'Indications']
    
    spice2indic_onegroup = cuisine_spice_dict[group]    
    indic2spice_temp = reverse_dict(spice2indic_onegroup)
    
    bi_graph, spice_topset, indic_botset = Bipartite_graph(spice2indic_onegroup, set_name)
    # row: bottom nodes - dict.values; column: top nodes - dict.keys - indicationsbot
    # top_set: indic; bot_set: spices
    adjmtx = bipartite.biadjacency_matrix(bi_graph, row_order=indic_botset, column_order=spice_topset).toarray()
    print('\n\nGroup Name: ', group)
    print('Adjacency Matrix Shape (#indication x #spice): ', adjmtx.shape)
    indic2spice_adjmtx.append(adjmtx)
    min_coverset_greedy = Min_cover_set_greedy(spice2indic_onegroup, indic2spice_temp)
    
    n2 = adjmtx.shape[1]
    c = np.hstack([np.ones(n2), [0]])


    min_coverset_optim = Min_cover_set(spice2indic_onegroup, spice_topset, indic_botset, adjmtx, c)
    print('\nSize of Min cover set matched: ', len(min_coverset_greedy) == len(min_coverset_optim))
    print('\nSet difference btw Greedy and Optimization approach: ', set(min_coverset_greedy) - set(min_coverset_optim))
    print('\nSymmetric difference difference btw Greedy and Optimization approach: ', set(min_coverset_greedy).symmetric_difference(set(min_coverset_optim)))# March 21,2020
    
    print('Min cover set by optimization:', len(min_coverset_optim), min_coverset_optim)
    print('\nMin cover set by greedy:', len(min_coverset_greedy), min_coverset_greedy)
    
    
    results_pd.loc[group, '#Spices'] = adjmtx.shape[1]
    results_pd.loc[group, '#Indications'] = adjmtx.shape[0]
    results_pd.loc[group, 'Min number of spices to cover indications'] = len(min_coverset_optim)
    results_pd.loc[group, 'Min set'] = min_coverset_optim
```
